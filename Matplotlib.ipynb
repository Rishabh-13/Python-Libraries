```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y =[99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x,y,color = "red", s =50)

plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')

plt.show()
```

```python
x = [2,4,6,8,10]
y = [1,5,8,10,14]

plt.scatter(x,y)
plt.show()
```

```python
## Bar Graph
course = np.array(["ML","AI","Python","Data Science" ])
no_students = np.array([3,4,6,5])

plt.bar(course,no_students, color= 'red', width = .2)
plt.xlabel('Courses')
plt.ylabel('Number of Students')
plt.title('Bar Graph')

plt.show()
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
df.Genre.value_counts()
```

```python
unique_cate = df.Genre.value_counts().index
```

```python
counts = df.Genre.value_counts().values
```

```python
plt.figure(figsize=(18,10)) ## To increase size of figure
plt.bar(unique_cate,counts)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Bar Graph')

plt.show()
```

```python
df1 = pd.read_csv('train.csv')
df1.head()
```

```python
df1.Survived.value_counts()
```

```python
labels = df1.Survived.value_counts().index
counts = df1.Survived.value_counts().values

plt.bar(labels,counts)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Bar Graph')

plt.show()
```
